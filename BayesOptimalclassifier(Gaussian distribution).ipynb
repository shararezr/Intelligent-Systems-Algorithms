```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import math
```

```python
train_data =  pd.read_csv("Train_Data.csv")
train_label =  pd.read_csv("Train_Labels.csv")
test_data =  pd.read_csv("Test_Data.csv")
test_label =  pd.read_csv("Test_Labels.csv")
train_data
```

### BayesOptimalclassifier (Gaussian distribution)

```python
def BayesOptimalclassifier(data,classes):
    
    #preprocessing
    data = (data-data.mean())/(data.std())
    n= data.shape[1]
    data['label'] = classes
    data['Gaussian'] = 0
    
    
    means = []
    labels = classes['0.0'].unique()
    labels.sort()
    priorK = []
    detcovsM = []
    invcovsM = []
    
    for l in labels:
    
        df = data[data['label'] == l]
        Q = df.shape[0]
        
        ##compute mean of each class
        mean = np.array(df.iloc[:,0:n].mean())
        means.append(mean)
        
        ##compute_prior_knowledge
        priorK.append(Q/data.shape[0])
        
        #compute Covariance Matrix of each class
        cov_matrix = np.zeros([n,n])
        for i in range(df.shape[0]):
            cov = (np.array(df.iloc[i,0:n]-mean)).reshape(1,n)
            cov_matrix = (np.dot(cov.transpose(),cov)).reshape(n,n)/(Q-1) + cov_matrix
               
        #compute det Covariance Matrix of each class
        det = np.linalg.det(cov_matrix)    
        detcovsM.append(det)
        
        #compute inverse Covariance Matrix of each class
        inv = np.linalg.inv(cov_matrix)
        invcovsM.append(inv)
        
        
        
    

    def predict(data):
        data = (data-data.mean())/(data.std())
        data = np.array(data)
        GaussiansPrior = np.zeros([data.shape[0],10])
        Gaussians = np.zeros([data.shape[0],10])
        print('prior_knowledge_probabilities')
        
        for l in range(10):
            #compute Gaussian
            X = data - means[l]
            crr = X.transpose()
            exp2 = np.zeros([data.shape[0],1])
            for i in range(data.shape[0]):
                exp1 = np.dot(X[i,:], invcovsM[l])
                exp2[i] = np.dot(exp1, crr[:,i])
                
            exp3 = ((2*math.pi)**(n/2))*math.sqrt(detcovsM[l])
            exp4 = np.exp((-1/2)*exp2)/exp3
            exp5 = (exp4*priorK[l])
            print(priorK[l])
            GaussiansPrior[:,l] = exp5[:,0]
            Gaussians[:,l] = exp4[:,0]
        Y_predict_P = GaussiansPrior.argmax(axis = 1)
        Y_predict = Gaussians.argmax(axis = 1)

        return Y_predict_P,Y_predict
       
    return predict
    
def accuracy(true, pred1, pred2):
    cnt1=0
    cnt2=0
    for i in range(len(Y_predict)):
        if(pred1[i] == true[i]):
            cnt1+=1
        if(pred2[i] == true[i]):
            cnt2+=1
    display("with prior knowledge:",cnt1/len(Y_predict)*100)
    display("without prior knowledge:",cnt2/len(Y_predict)*100)

def compute_confusion_matrix(true, pred1, pred2):

    K = len(np.unique(true)) # Number of classes 
    result1 = np.zeros([K, K])
    result2 = np.zeros([K, K])

    for i in range(len(true)):
        result1[true[i],pred1[i]] += 1
        result2[true[i],pred2[i]] += 1
        
    classes = np.unique(true)
    total_classes = len(classes)
    level_labels = [total_classes*[0], list(range(total_classes))]

    cm_frame1 = pd.DataFrame(data=result1,columns=pd.MultiIndex(levels=[['Predicted_label:'],\
                                                                  classes],codes=level_labels),\
                        index=pd.MultiIndex(levels=[['True_label:'], classes], codes=level_labels))
    
    
    cm_frame2 = pd.DataFrame(data=result2,columns=pd.MultiIndex(levels=[['Predicted_label:'],\
                                                                  classes],codes=level_labels),\
                        index=pd.MultiIndex(levels=[['True_label:'], classes], codes=level_labels))
    
    
    display("with prior knowledge:",cm_frame1)   
    display("without prior knowledge:",cm_frame2)
 
```

### Test Data:

```python
func = BayesOptimalclassifier(train_data,train_label)
Y_predict_P,Y_predict = func(test_data )
```

### (Confusion Matrix) and (Accuracy):

```python
Truelabel = np.array(test_label['2.0']).astype('int32') 
accuracy(Truelabel, Y_predict_P, Y_predict)
compute_confusion_matrix(Truelabel, Y_predict_P, Y_predict)
```

Because the probabilities of prior knowledge are almost the same for different classes, it has no effect on the final answer.
